### HEA Distribution Study
***

**Experiment**: Samples of HEA ranging from DP 100-575 in steps of 25. Samples were polymerized under the lamp at 5 inches for 4 hours. HEA polymers were made using CTA 6

**Comments:** I think this would look wonderful with a few polymers being tested. Obvious monomers would be NAM, DMA, and NIPAM in addition to HEA. Given our resolution on the high throughput column, the seperation of the Mn and Mw really sticks out here. That being said, the Mw stays very very consistent and regular with the theory.



```python
DF_Temp = df.loc[df['Study'] == 'KF_01_04']

HEA_Mn = 116
Hea_th = DF_Temp['DP']*HEA_Mn

fig, ax = plt.subplots(figsize = (9,6))
ax.scatter(DF_Temp['DP'],DF_Temp['Mn_HT'],color = 'blue')
ax.scatter(DF_Temp['DP'],DF_Temp['Mw_HT'],color = 'black')
ax.scatter(DF_Temp['DP'],Hea_th,color = 'red')
ax.set_title('HEA DP 100-575')
ax.set_xlabel('Degree of Polymerization')
ax.set_ylabel('Molecular Weight')
ax.legend(['Mn','Mw','Mn_Theory'])
```
