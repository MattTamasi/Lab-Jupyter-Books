####       Reversible addition-fragmentation reaction (RAFT) polymerization techniques have been of growing interest and have been adopted for the exploration of a wide range of biological technologies. These RAFT polymerization techniques have offered researchers precise control over molecular weight and polydispersity which in turn has allowed investigation of complex architectures and bioactive functionality. While current applications of RAFT for biotechnologies are heavily focused on bioconjugation, supramolecular assemblies, and surface modifications, the exploration of rationally designed polymers for protein dependent biointerfaces has remained insufficiently challenged due to its inherent complexity. Here we describe the potential of adapting a recently discovered porphyrin-catalysed photoinduced electron transfer-reversible addition-fragmentation chain-transfer (PET-RAFT) polymerization to a Microlab STAR liquid handling robot to create a system for high throughput synthesis of customizable polymers for a variety of biological applications.

#### Monomer vs Molecular Weight

```python

def PlotMn(Monomer):
    df3 = df.loc[df['Monomer 1'] == Monomer]

    Array2 =[]

    x = ['DP 100', 'DP 200', 'DP 400']
    y = df3.loc[df3['CTA'] == 'CTA 2']
    y = y['Mn_HT']
    y2 = df3.loc[df3['CTA'] == 'CTA 3']
    y2 = y2['Mn_HT']
    y3 = df3.loc[df3['CTA'] == 'CTA 4']
    y3 = y3['Mn_HT']
    y4 = df3.loc[df3['CTA'] == 'CTA 6']
    y4 = y4['Mn_HT']
    y5 = df3.loc[df3['CTA'] == 'CTA 7']
    y5 = y5['Mn_HT']
    y6 = df3.loc[df3['CTA'] == 'CTA 8']
    y6 = y6['Mn_HT']


    layout = go.Layout(
        title='Polymer Molecular Weights vs CTA for: Methyl Acrylate',
    )


    CTA2 = go.Bar(
        x=x,
        y=y,
        name = 'CTA 2',
        text=y,
        textposition = 'auto',
        marker=dict(
            color='rgb(255,2,5)',
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
            ),
        opacity=0.6
    )

    CTA3 = go.Bar(
        x=x,
        y=y2,
        name = 'CTA 3',
        text=y2,
        textposition = 'auto',
        marker=dict(
            color='rgb(158,200,0)',
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
            ),
        opacity=0.6
    )

    CTA4 = go.Bar(
        x=x,
        y=y3,
        name = 'CTA 4',
        text=y3,
        textposition = 'auto',
        marker=dict(
            color='rgb(158,20,225)',
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
            ),
        opacity=0.6
    )

    CTA6 = go.Bar(
        x=x,
        y=y4,
        name = 'CTA 6',
        text=y4,
        textposition = 'auto',
        marker=dict(
            color='rgb(158,0,225)',
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
            ),
        opacity=0.6
    )

    CTA7 = go.Bar(
        x=x,
        y=y5,
        text=y5,
        name = 'CTA 7',
        textposition = 'auto',
        marker=dict(
            color='rgb(0,150,0)',
            line=dict(
                color='rgb(8,48,107)',
                width=1.5),
            ),
        opacity=0.6
    )

    CTA8 = go.Bar(
        x=x,
        y=y6,
        name = 'CTA 8',
        text=y6,
        textposition = 'auto',
        marker=dict(
            color='rgb(0,255,225)',
            line=dict(
                color='rgb(255,255,255)',
                width=1.5),
            ),
        opacity=0.6
    )


    data = [CTA2,CTA3,CTA4,CTA6,CTA7,CTA8]
    fig = go.Figure(data=data, layout=layout)
    iplot(fig, filename='grouped-bar-direct-labels')
    
    
interact(PlotMn, Monomer=['MMA','HEA','EtA','Sty','BA','NAM','DMA']);
```

### HEA Distribution Study
***

**Experiment**: Samples of HEA ranging from DP 100-575 in steps of 25. Samples were polymerized under the lamp at 5 inches for 4 hours. HEA polymers were made using CTA 6

**Comments:** I think this would look wonderful with a few polymers being tested. Obvious monomers would be NAM, DMA, and NIPAM in addition to HEA. Given our resolution on the high throughput column, the seperation of the Mn and Mw really sticks out here. That being said, the Mw stays very very consistent and regular with the theory.



```python
DF_Temp = df.loc[df['Study'] == 'KF_01_04']

HEA_Mn = 116
Hea_th = DF_Temp['DP']*HEA_Mn

fig, ax = plt.subplots(figsize = (9,6))
ax.scatter(DF_Temp['DP'],DF_Temp['Mn_HT'],color = 'blue')
ax.scatter(DF_Temp['DP'],DF_Temp['Mw_HT'],color = 'black')
ax.scatter(DF_Temp['DP'],Hea_th,color = 'red')
ax.set_title('HEA DP 100-575')
ax.set_xlabel('Degree of Polymerization')
ax.set_ylabel('Molecular Weight')
ax.legend(['Mn','Mw','Mn_Theory'])
```
