Note: These polymers are constructed of `MMA-PEGMA-HPMA-DMAPMA (+)`

```python
data = pd.read_excel(r'C:\Users\Matt Tamasi\Desktop\Python Codes\Lipase Trial.xlsx')
```

```python
plt.figure(figsize=(12, 6))
ax = sns.pointplot('Charge', 'Absorbance', hue = 'Exp',
    data=data, dodge=True, join=False,palette = 'Set2')
text = ax.set_title('Lipase Activity vs Cationic Charge')
```
